## LIKE işlemleri
* 'a%' : “a” ile başlayan değerler
* '%a' : “a” ile biten değerler
* %or%' : herhangi bir yerinde “or” olan değerler
* '_r%' : ikinci sirasında “r” olan değerler
* 'a%%' : “a” ile başlayıp en az 3 karakterden oluşan değerler
* 'a%o' : “a” ile başlayıp “o” ile biten değerler

```python
%%sql
select * from customers limit 10
```

```python
%%sql
select * from customers where CustomerName like "%or%" limit 10
```

```python
%%sql
select * from customers where CustomerName like "%a" limit 10
```

```python
%%sql
select * from customers where CustomerName like "a%" and City like "%in" 
```

```python
%%sql
select * from Products
left outer join Categories on Products.CategoryID=Categories.CategoryID
```

```python
%%sql
select Products.CategoryID, Products.ProductName, Categories.CategoryName from Products
left outer join Categories on Products.CategoryID=Categories.CategoryID
```

```python
%%sql
select pr.CategoryID, pr.ProductName, cat.CategoryName from Products as pr
left outer join Categories as cat on pr.CategoryID=cat.CategoryID limit 10
```

## SQL ve Pandas DataFrame Operasyonları

```python
import pandas as pd
```

```python
%sql SELECT COUNT(CustomerID) AS MUSTERI_SAYISI, Country FROM Customers GROUP BY Country ORDER BY MUSTERI_SAYISI DESC;
```

```python
df = %sql SELECT COUNT(CustomerID) AS MUSTERI_SAYISI, Country as ULKE FROM Customers GROUP BY Country ORDER BY MUSTERI_SAYISI DESC;
```

```python
type(df)
```

```python
df=pd.DataFrame(df)
df.head()
```

```python
df.plot.barh(x="ULKE",y="MUSTERI_SAYISI")
```
